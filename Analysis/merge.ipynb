```python
import pandas as pd
```

```python
df = pd.read_csv("full.csv")
```

```python
df.head()
```

```python
df["duration"] = df["duration"].apply(lambda x : sum(mul * int(x) for mul, x in zip([60, 1], x.split(":"))))
df["duration"].head()
```

```python
key_dict = {
	"0": "C",
	"1": "C♯",
	"2": "D",
	"3": "D♯",
	"4": "E",
	"5": "F",
	"6": "F♯",
	"7": "G",
	"8": "G♯",
	"9": "A",
	"10": "A♯",
	"11": "B"
}
df["key"] = df["key"].apply(lambda x : key_dict[str(x)])
df["key"].head()
```

```python
len(df["key"].unique())
```

```python
tsig = ["minor", "major"]
df["mode"] = df["mode"].apply(lambda x : tsig[x])
```

```python
df["mode"].nunique()
```

```python
df["time_signature"] = df["time_signature"].apply(lambda x: "{}/4".format(x))
```

```python
df["time_signature"].unique()
```

```python
df.to_csv("final_converted.csv", sep="\t", index=False)
```
