```python
import pandas as pd
```

```python
yt = pd.read_csv("../Scrape/youtube/youtube_rescrape.csv")
```

```python
sp_playcount = pd.read_csv("../Scrape/Spotify/playcount.csv")
sp_playcount.head()
```

```python
sp_playcount_sub = sp_playcount[["song_id", "playcount"]]
sp_playcount_sub.head()
```

```python
sp_feature = pd.read_csv("../Scrape/Spotify/spotify-playlist.csv", sep="\t")
sp_feature.head()
```

```python
sp_feat_sub = sp_feature.drop(columns=["Unnamed: 0", "artist_id", "album_id"])
sp_feat_sub.head()
```

```python
print("len of spotify feature: {}".format(len(sp_feat_sub)))
print("len of spotify playcount: {}".format(len(sp_playcount)))
```

```python
sp = sp_feat_sub.join(sp_playcount_sub.set_index("song_id"), on="track_id")
```
