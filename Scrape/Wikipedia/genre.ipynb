```python
genre_dict = {}

genre_dict["rock"] = rock_keywords
genre_dict["pop"] = pop_keywords
genre_dict["electronic-dance"] = edm_keywords
genre_dict["hip hop"] = hip_hop_keywords
# genre_dict["classical"] = classical_keywords
genre_dict["jazz"] = jazz_keywords
genre_dict["folk"] = folk_keywords
genre_dict["reggae"] = reggae_keywords
# genre_dict["blues"] = blues_keywords
genre_dict["r&b"] = r_and_b_keywords
genre_dict["metal"] = metal
genre_dict["country"] = country
# genre_dict["latin"] = latin
```

```python
import numpy as np

def check_genre(sub_genre):
	if type(sub_genre) == float:
		return False
	for big_genre in genre_dict.keys():
		for keyword in genre_dict[big_genre]:
			if keyword in sub_genre:
				return False
	return True

def convert_genre(sub_genre):
	if type(sub_genre) == float:
		return sub_genre
	for big_genre in genre_dict.keys():
		for keyword in genre_dict[big_genre]:
			if keyword in sub_genre:
				return big_genre
	print("wtf {}".format(sub_genre))

def replace_with_nan(sub_genre):
	if sub_genre in replace_nan_list:
		return np.NaN
	return sub_genre
```

```python
replace_nan = df.copy()
replace_nan["genres"] = df["genres"].apply(replace_with_nan)
print(replace_nan.isna().sum())
replace_nan["genres"] = replace_nan["genres"].apply(convert_genre)
replace_nan.head()
replace_nan = replace_nan.drop_duplicates(ignore_index=True)
replace_nan.to_csv("genre_converted.csv", index=False)
```
