```python
for i, row in enumerate(soup.tbody):
	if i % 2 == 1:
		print(i, row)
		break
```

```python
for i, row in enumerate(soup.tbody):
	if i % 2 == 1:
		print(row.findAll('td'))
		break
```

```python
for i, row in enumerate(soup.tbody):
	if i % 2 == 1:
		result = row.findAll('td')
		for e in result:
			print(e)
		break
```

```python
for i, row in enumerate(soup.tbody):
	if i % 2 == 1:
		info = row.findAll('td')
		link = row.a['href']
		print(link)
		l = []
		for e in info:
			l.append(e.text)
		print(l)
		break
```

```python
for i, row in enumerate(soup.tbody):
	if i % 2 == 1:
		info = row.findAll('td')
		id = row.a['href'].split('/')[1][:-5]
		print(id)
		l = []
		for e in info:
			l.append(e.text)
		print(l)
		break
```

```python

for i, row in enumerate(soup.tbody):
	if i % 2 == 1:
		info = row.findAll('td')
		id = row.a['href'].split('/')[1][:-5]
		print(id)
		l = [id]
		for e in info:
			l.append(e.text)
		print(l)
		break
```

```python
import csv
```

```python
with open("kworb.csv", 'w') as f:
	writer = csv.writer(f)
	writer.writerow(["id", "title", "view"])
	for i, row in enumerate(soup.tbody):
		if i % 2 == 1:
			info = row.findAll('td')
			id = row.a['href'].split('/')[1][:-5]
			print(id)
			l = [id]
			for e in info:
				l.append(e.text)
			print(l)
			writer.writerow(l[:-1])
```
